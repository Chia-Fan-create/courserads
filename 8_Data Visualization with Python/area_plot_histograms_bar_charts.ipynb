```python
# Import Primary Modules
import numpy as np
import pandas as pd

# Use the inline backend to generate the plots within the browser
%matplotlib inline

# Import Matplotlib Modules
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# Print the version of Matplotlib
print("Matplotlib version: ", mpl.__version__)  # >= 2.0.0
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

print('Data read into a pandas dataframe!')
```

```python
df_can.head()
```

```python
# print the dimensions of the dataframe
print(df_can.shape)
```

```python
df_can.set_index('Country', inplace=True)

# Let's view the first five elements and see how the dataframe was changed
df_can.head()
```

```python
print('data dimensions:', df_can.shape)
```

```python
# finally, let's create a list of years from 1980 - 2013
# this will come in handy when we start plotting the data
years = list(map(str, range(1980, 2014)))

years
```

Area Plots

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df_can.head()
# df_top5

# transpose the dataframe
df_top5 = df_top5[years].transpose()

df_top5.head()
# df_top5
```

```python
# let's change the index values of df_top5 to type integer for plotting
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',
             stacked=False,
             figsize=(20, 10))

plt.title('Immigration Trend of Top 5 Countries to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_top5.plot(kind='area',
             alpha=0.25, # 0-1, default value alpha = 0.5
                stacked=True,
                figsize=(20, 10))

plt.title('Immigration Trend of Top 5 Countries to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
# Option1
df_top5.plot(kind='area', alpha = 0.35, stacked=False, figsize=(20, 10))
plt.title('Immigration Trend of Top 5 Countries to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
# Option2 : preferred option with more flexibility
ax = df_top5.plot(kind='area', alpha = 0.35, figsize=(20, 10))
ax.set_title('Immigration Trend of Top 5 Countries to Canada')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
df_least = df_can.tail(5)
# transpose the dataframe
df_least = df_least[years].transpose()

# change the index values of df_least to type integer for plotting
df_least.index = df_least.index.map(int)
df_least.plot(kind='area',
             alpha=0.45, # 0-1, default value alpha = 0.5
                figsize=(20, 10))

plt.title('Immigration Trend of 5 Least Contributing Countries to Canada')
plt.ylabel('Number of Immigrants')  
plt.xlabel('Years')
plt.show()
```

```python
ax = df_least.plot(kind='area', alpha = 0.55, stacked=False, figsize=(20, 10))
ax.set_title('Immigration Trend of 5 Least Contributing Countries to Canada')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
# get the x-tick values
count, bin_edges = np.histogram(df_t, 15)

# un-stacked histogram
df_t.plot(kind='hist', bins=15, figsize=(10, 6), alpha=0.6, xticks=bin_edges, color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden to Canada (1980-2013)')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10   #  first bin value is 31.0, adding buffer of 10 for aesthetic purposes 
xmax = bin_edges[-1] + 10  #  last bin value is 308.0, adding buffer of 10 for aesthetic purposes

# stacked Histogram
df_t.plot(kind='hist',
          figsize=(10, 6), 
          bins=15,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

Bar Charts(DataFrame)

```python
# step1: get the data
df_iceland = df_can.loc['Iceland', years]
df_iceland.head()
```

```python
# plot the dataframe
df_iceland.plot(kind='bar', figsize=(10, 6))

plt.title('Icelandic Immigration to Canada from 1980 to 2013')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_iceland.plot(kind='bar', figsize=(10, 6), rot=90)
plt.title('Icelandic Immigration to Canada from 1980 to 2013')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate arrow
plt.annotate('', # s:str. Will leave it blank for no text
             xy=(32, 70), # place head of the arrow at point (year 2012, population 70)
             xytext=(28, 20), # place base of the arrow at point (year 2005, population 20)
             xycoords='data', # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)) # properties of the arrow

plt.show()
```

```python
df_iceland.plot(kind='bar', figsize=(10, 6), rot=90)

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('',  # s: str. will leave it blank for no text
             xy=(32, 70),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis',  # text to display
             xy=(28, 30),  # start the text at at point (year 2008 , pop 30)
             rotation=72.5,  # based on trial and error to match the arrow
             va='bottom',  # want the text to be vertically 'bottom' aligned
             ha='left',  # want the text to be horizontally 'left' algned.
             )

plt.show()
```

Question: Using the scripting later and the df_can dataset, create a horizontal bar plot showing the total number of immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total immigrant count.

```python
# step1: get the data
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)
# get the top 15 entries
df_top15 = df_can['Total'].tail(15)
df_top15
```

```python
# plot the dataframe
from turtle import color


df_top15.plot(kind='barh', figsize=(12, 12), color = 'teal')

plt.title('Immigration from Top 15 Countries to Canada in 2013')
plt.ylabel('Number of Immigrants')

# annotate value labels to each country
for index, value in enumerate(df_top15.values):
    label = format(int(value), ',')  # format the value with commas
    plt.annotate(label, 
                 xy = (value - 47000 , index - 0.1),
                 xytext= ( value , index),
                 xycoords = 'data',
                 color = 'black',
                 va = 'center'
                )  
plt.show()
```
